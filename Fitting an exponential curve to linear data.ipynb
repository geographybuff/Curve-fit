```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
x = range(20)
y = np.array([x[i]/2 + 3*np.random.uniform() for i in range(20)])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def f(x, a):
    return np.exp(x*a)
coefficient = opt.curve_fit(f, x, y)
```

```python
plt.scatter(x,y)
l = np.linspace(0, 19, 500)
ft = [np.exp(l[i]*coefficient[0][0]) for i in range(len(l))]
plt.plot(l, ft, "-r")
plt.show()
```
